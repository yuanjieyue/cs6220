```python
from sklearn import preprocessing as pp
```

```python
df.boxplot(grid=False,figsize=(48,12))
```

```python
# 1. Standardize the imputed feature data so that the values of each numeric feature are standard normally distributed
for col in df:
    df[col] = pp.scale(df[col], axis=0, with_mean=True, with_std=True, copy=True)
```

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
```

```python
X=df
y=df_class
```

```python
# 1. To get an idea of their relative importance, estimate the mutual information between 
# the numeric features and the class column, ‚ÄòSalePrice‚Äô
mi = mutual_info_classif(X, y)
```

```python
mi
```

```python
mi_infos=pd.DataFrame(mi)
```

```python
# rename the column as score
mi_infos.rename(columns={0: 'score'},inplace=True)
```

```python
# add the original feature name as a new column.
mi_infos['feature'] = pd.DataFrame(X.columns)
```

```python
# sort the new dataframe by score in descending order.
mi_infos.sort_values(by='score',ascending=False)
```

```python
# 2. What are the top 5 numeric features ranked by mutual information?
```

```python
# Select the k best features.
selector = SelectKBest(mutual_info_classif, k=5)
selector.fit(X, y)

# Get indices of columns to keep.
idxs_selected = selector.get_support(indices=True)

# Create new DataFrame with only desired columns.
X_new = X.iloc[:, idxs_selected]

# Show the top 5 features.
X_new.head()
```
