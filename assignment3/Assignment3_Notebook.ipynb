````python
%%markdown
# 1. Online Retail Market Basket Analysis
````

````python
pd.set_option("max_colwidth", 150)
f = "https://github.com/cs6220/cs6220.spring2019/raw/master/data/Online%20Retail.xlsx" 
df = pd.read_excel(f) 
````

````python
df.head()
````

````python
df.count()
````

````python
basket = (df[df["Country"] == "United Kingdom"]
          .groupby(["InvoiceNo", "Description"])["Quantity"]
          .sum().unstack().reset_index().fillna(0)
          .set_index("InvoiceNo")) # transform transactions into baskets of items 
````

````python
basket_sets = basket.applymap(lambda x: 1 if x >=1 else 0) # convert counts to booleans
````

````python
basket_sets.head()
````

````python
%%markdown
## 1.1 Frequent Itemset Generation
````

````python
#1. What are the top 5 1-itemsets with the highest support?
````

````python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True, verbose=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets.head()
````

````python
top_five_1_itemsets = frequent_itemsets[frequent_itemsets['length'] == 1].sort_values(by='support',ascending=False).head()
top_five_1_itemsets
````

````python
#2. What are the top 5 2-itemsets with the highest support?
````

````python
top_five_2_itemsets = frequent_itemsets[frequent_itemsets['length'] == 2].sort_values(by='support',ascending=False).head()
top_five_2_itemsets
````

````python
#3. What is the highest support value for the 1-itemsets?
````

````python
top_five_1_itemsets.head(1)
````

````python
#4.What is the highest support value for the 2-itemsets?
````

````python
top_five_2_itemsets.head(1)
````

````python
%%markdown
## 1.2 Association Rule Generation
````

````python
#1. What are the top 5 association rules?
````

````python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
````

````python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules["consequent_len"] = rules["consequents"].apply(lambda x: len(x))
rules.sort_values(by='confidence',ascending=False).head()
````

````python
# 1) first try min_support = 0.6
````

````python
frequent_itemsets = apriori(df_adult_dummy, min_support=0.6, use_colnames=True, verbose=True)
frequent_itemsets.head()
````

````python
confidence_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
confidence_rules['antecedents_len'] = confidence_rules['antecedents'].apply(lambda x: len(x))
confidence_rules['consequents_len'] = confidence_rules['consequents'].apply(lambda x: len(x))
confidence_rules['len'] = confidence_rules['consequents_len'] + confidence_rules['antecedents_len']
confidence_rules.sort_values(by='confidence',ascending=False).head(3)
````

````python
lift_rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.9)
lift_rules['antecedents_len'] = lift_rules['antecedents'].apply(lambda x: len(x))
lift_rules['consequents_len'] = lift_rules['consequents'].apply(lambda x: len(x))
lift_rules['len'] = lift_rules['consequents_len'] + lift_rules['antecedents_len']
lift_rules.sort_values(by='lift',ascending=False).head(3)
````

````python
%%markdown
When the threshold of support set to be 0.6, we could see several interesting things from the results we get above:
1. All of the top rules are 1 -> 1 format. 


2. For the top 3 confidence rules and lift rules, there are 6 antecedents and 6 consequents, among which 
native-country_ United-States appear 6 times followed by race_ White 3 times. We could say that most of the the top 
rules are around race_ White and native-country_ United-States. This could be explained by
what we have calculate before that the ratio of White in race is around 85% and United-States in native-country is 
around 89%, which can explain the rules with metric 'confidence' and 'lift' so well. 
Because for every antecedents combinations, these two features would most likely to appear, which will centainly raise
up the support of those rules, now if the support of the antecedent is not that high, then the final confidence will 
be higher. 


3. the rule with the highest confidence and the rule with the highest lift is the the one,
    race_ White --> native-country_ United-States: The ratios are exactly the 
antecedents support and consequents support. We could emulate the calculating process using the formular as below:
```
    support(A -> C)    = support(A U C) = 0.78
    confidence(A -> C) = support(A -> C) / support(A) 
                       = 0.78 / 0.85
                       ≈ 0.92
    lift(A -> C)       = confidence(A -> C) / support(C)
                       = 0.92 / 0.89
                       ≈ 1.03
```
````
